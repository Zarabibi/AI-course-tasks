```python
import torch
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

# Load YOLOv5 model (dataset)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Load an image
img_path = '/content/00000003.jpg'  # Replace with your test image path
img = Image.open(img_path)

# Perform detection
results = model(img)

# Print results
results.print()  # prints detected objects

# Show results
results.show()

# Save results to file
save_path = Path('runs/detect/exp')
results.save()  # saves annotated image to 'runs/detect/exp'

print(f'Detection results saved to {save_path.resolve()}')
```
