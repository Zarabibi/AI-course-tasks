```python
from sentence_transformers import SentenceTransformer, util
```

```python
# Load pre-trained sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
import pandas as pd
data = pd.read_csv('/content/Resume.csv')
```

```python
file_path = '/content/Resume.csv'
try:
    with open(file_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            # You can add checks here to see if lines look malformed
            pass # Just reading lines to check if it completes
    print("File read successfully line by line.")
except Exception as e:
    print(f"Error reading file at line {i+1}: {e}")
```

```python
# Job description text
job_description = """
We are looking for a HR MANAGER , HR CLEARK and HR ASSISTENR.
"""
```

```python
# Encode job description and resumes
job_desc_embedding = model.encode(job_description, convert_to_tensor=True)
resume_embeddings = model.encode(data['Resume_str'], convert_to_tensor=True)
```

```python
# Compute cosine similarities between job description and each resume
cosine_scores = util.pytorch_cos_sim(job_desc_embedding, resume_embeddings)[0]
```

```python
# Rank resumes by similarity score (higher is better)
ranked_indices = cosine_scores.argsort(descending=True)
```

```python
print("Top 5 resumes ranked by relevance to job description:")
# Convert ranked_indices tensor to a list of integers
ranked_indices_list = ranked_indices.tolist()

# Limit to top 5 resumes
top_5_indices = ranked_indices_list[:5]

for rank, idx in enumerate(top_5_indices):
    print(f"{rank + 1}. Resume {idx + 1} — Score: {cosine_scores[idx]:.4f}")
    print(f"Resume Text: {data['Resume_str'].iloc[idx]}")
    print(f"Job Description: {job_description}\n")
```
